```python
def plot_training_metrics(training_loss,validation_loss,iou,pixel_accuracy):
    epochs = range(1, len(training_loss) + 1)
    plt.figure(figsize=(12, 8))

    # Plot IOU
    plt.plot(epochs, iou, 'b', label='IOU')

    # Plot training and validation losses
    plt.plot(epochs, training_loss, 'r', label='Training Loss')
    plt.plot(epochs, validation_loss, 'g', label='Validation Loss')

    # Plot pixel accuracy
    plt.plot(epochs, pixel_accuracy, 'y', label='Pixel Accuracy')

    plt.title('Training Metrics')
    plt.xlabel('Epochs')
    plt.ylabel('Metrics')
    plt.legend()
    plt.grid(True)
    plt.show()
```

```python
std_save = "./models/segnet/segnet_standard_0.5"

std_training_loss = torch.load(os.path.join(std_save,"train_loss.pt"))
std_validation_loss = torch.load(os.path.join(std_save,"val_loss.pt"))
std_iou = torch.load(os.path.join(std_save,"iou.pt"))
std_pixel_accuracy = torch.load(os.path.join(std_save,"pixel_acc.pt"))
```

```python
plot_training_metrics(std_training_loss, std_validation_loss, std_iou, std_pixel_accuracy)
```

```python
dsc_save = "./models/segnet/segnet_dsc_0.5"

dsc_training_loss = torch.load(os.path.join(dsc_save,"train_loss.pt"))
dsc_validation_loss = torch.load(os.path.join(dsc_save,"val_loss.pt"))
dsc_iou = torch.load(os.path.join(dsc_save,"iou.pt"))
dsc_pixel_accuracy = torch.load(os.path.join(dsc_save,"pixel_acc.pt"))
```

```python
plot_training_metrics(dsc_training_loss, dsc_validation_loss, dsc_iou, dsc_pixel_accuracy)
```
